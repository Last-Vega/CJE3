### ■ Pythonのヒント：インデント

`Python`のプログラムを書く際に注意しなければならないのは、**行のインデント（字下げ）**です。

`Ruby`には`for`ループなどの処理ブロックの終わりを告げる`end`がありましたが、`Python`にはありません。その代わりインデントがブロックの終わりを示す役割をしています。

例えば、上のコードセルの`for`ループ処理では、ループ内の手続きである

> `print(line)`

がある行の頭にインデントが入っています。つまり、この`for`ループの処理内容には、それ以降のインデントのある行が含まれる、ということを意味しています。

`Python`ではこのようにループや条件文の中身を**一貫したスタイル**でインデントしないとエラーがでます。

試しに、上のコードセルを修正して、`print(line)`の頭にある空白を削除して、再実行してみましょう。

すると

> `IndentationError`

と表示されたエラーが出ると思います。このエラーがでたらプログラムのどこかで**インデントが正しく行われていない**という意味ですので、修正しましょう。

#### なぜインデントが正しくないとエラーが出るのか？

冒頭で述べた通り、`Python`ではインデントを使って処理ブロックの開始と終了地点（**スコープ**といいます）を判断しています。上記の`for`ループの場合も、繰り返し行う処理が続く範囲は、**インデントされた行がどこまで続くのか**、によって判断されます。したがって、インデントがない行が現れると、プログラマーはまだループ処理を続けたいと考えても、`Python`はループ処理が終わり、次の別処理を命じられたと解釈してしまいます。

#### **一貫したスタイル**って何？

一貫したインデントを使ってプログラムを書くには、大きく分けて二つ注意する必要があります。

1. インデントは、タブ文字か空白文字のどちらかで統一する
1. 入れ子の層が同じ深さの処理は、同じインデントの長さをもつ

1点目のおすすめ対処方法は、**常にタブ（tab）キーを使って**インデントを入力することです。タブ文字は大抵空白文字の2文字か4文字分の長さの空白を入力するように設定されているので、何度もスペースキーを押すよりも効率的です。

> `Jupyter Lab`では、普通に`Enter`キーで改行すれば、自動的に適切な量のタブ文字（＝インデント）が入力されるように設定されています。

2点目に関しては、プログラムの中身を理解して、今書いている処理が、何層目の入れ子になっているかに注意を払う必要があります。慣れないうちはエラーがよく出るかもしれませんが、インデントが正しくできているプログラムは、一般的に人間にとって**読みやすいプログラム**でもあるので、一貫したインデントができるような習慣を身につけましょう。

### 文を単語に分ける

さて、無事にファイルの中身を1行ずつ表示するプログラムが書けました。次の処理は、読み込んだ行を**単語に分ける**ことです。

講義で説明したように、索引付け処理とは、本の後ろにある索引表（と同様の役割をもつもの）をウェブページを対象に作成することです。本の索引表も大抵の場合、本の中で使われたキーワードが使われていると思います。同様に、検索システムでも読み込んだファイル（ウェブページ）の索引を、文単位ではなく、単語単位で作成します。

> 文を単語に分けることを**分かち書き**と呼び、日本語や中国語など単語の間に空白のない言語では、よく使われる手法です。  
> また、分かち書き処理によって生成される文字列は、正確には**単語だけではない**のですが、本演習ではその辺りを区別せずに進めます。  
> さらに、ここで分かち書き処理と呼んでいるものは、正確には**形態素解析**と呼ばれるものですが、本演習では分かち書きで進めます。

読み込んだ行を分かち書きするには、以下のようにします。

````python
from janome.tokenizer import Tokenizer
t = Tokenizer()

DATA = "../data"
f = open(DATA + '/sample_doc1.txt', 'r')

for line in f:
    tokens = t.tokenize(line)
    
    for token in tokens:
        print(token.surface)
````

---
上のプログラムは、

> 1: 分かち書きをするPythonライブラリ`janome.tokenizer`から、分かち書きクラス`Tokenizer`を読み込む  
> 2: 分かち書きクラス`Tokenizer`を初期化し、クラスのショートカット（正しくは・・・略）を変数`t`に格納する  
> 4: [説明済み]  
> 5: [説明済み]  
> 7: [説明済み]  
> 8: 変数`line`に格納された行を、`tokenize`メソッドで分かち書きして、その結果を変数`tokens`に格納  
> 10: 変数`tokens`に格納された最初の単語の分かち書きデータへのショートカットを、変数`token`に格納。`for`メソッドを使って、以下の処理を単語がなくなるまで繰り返す。  
> 11: 変数`token`が参照する分かち書きデータから、`surface`メソッドを使って**単語本体のみ**を表示

という処理をしています。

#### １行ずつしっかりと理解を！

処理内容は沢山ありますが、あせらずにゆっくりと順に読んで、しっかりと理解してください。特に`for`ループは見かけよりも**多くの処理をしています**ので、これまで曖昧に理解していた人は、上の説明を何度もよく読み返して、その処理を正確に説明できるようになりましょう。

**ライブラリ・クラス・変数・メソッド**などのキーワードの区別にも注意を払いましょう。 

また、**インデントが揃っている**ことにも注意してください。インデントが揃っているおかげで、7行目からはじまる`for`ループは11行目までの処理を繰り返す、10行目から始まる`for`ループは11行目までの処理を繰り返す、という流れが一目でわかります。


---
それでは、上のコードセルを実行してみましょう。

### 単語の出現回数を数える

無事にファイルの中身を分かち書きすることができたので、次に単語の出現回数を数えるプログラムを作成しましょう。単語の出現回数は検索システムのランキングに不可欠な情報の一つです。

単語の出現回数を数えるときには、**辞書オブジェクト**を使うと便利です。辞書オブジェクトは、`Ruby`の連想配列（ハッシュ）と似ていて、**キー**と**値**がセットになって一つのレコードを表現します。

辞書オブジェクト`dict`の例を見ていきましょう。

````python
dict = {"我輩":64, "猫":75, "書生":82} # 辞書オブジェクトの作成
print(dict['我輩']) # キー'我輩'の値の出力
dict['我輩'] += 1 # キー'我輩'の値に1を加える
dict['柴犬'] = 1 # 新しいレコードの追加

# 全体の出力
for key in dict:
    print(key + '\t' + str(dict[key]))
````

---
上のプログラムは、

> 1: 辞書オブジェクト`dict`を作成し、レコードを3件登録  
> 2: `dict`オブジェクト内の**我輩**をキーとするレコードの値を出力  
> 3: **我輩**をキーとするレコードの値に1を加える  
> 4: **柴犬**をキー、**1** を値とする、新しいレコードの追加  
> 7: `for`メソッドを使って、`dict`オブジェクト内のキーを順番に変数`key`に格納。以下の処理をキーがなくなるまで繰り返す。  
> 8: `print`メソッドを使って、変数`key`と`dict`オブジェクト内で`key`が参照する値を出力。

という処理をしています。

注意点は8行目です。`dict[key]`が参照する変数は**数値オブジェクト**であり、**文字列オブジェクト**ではないので、そのまま直接出力できません。そこで、`str`メソッドを使って、`dict[key]`を一旦文字列オブジェクトに変換しています。

> こういったオブジェクトの種類（型）を変更することを、**キャストする**ということがあります。

それでは実行して、辞書オブジェクトがどのようにキーと値を格納するのか、**我輩**をキーとするレコードの値が変更されているか、などを確認してみましょう。

今回の例では、入力順序と出力順序が同じかもしれませんが、辞書オブジェクトは、**入力順序を保持しない**ので注意が必要です。

#### 辞書オブジェクトの出力結果をソートする

辞書オブジェクトを使用するプログラムでは、辞書オブジェクトの中身を出力するときに、内容をあいうえお順などで並べたいことがあります。この並べ替えのことを**ソート**と呼びます。ソート方法は複数ありますが、ここでは一番単純なあいうえお順（あるいはアルファベット順）でのソートを実装しましょう。

やり方は簡単で、辞書オブジェクト`dict`を`for`ループに渡す前に、辞書オブジェクトのキーをあいうえお順に並べたものを出力する`sorted`メソッドを使うだけです。以下のプログラムの7行目に着目してください。

````python
dict = {"我輩":64, "猫":75, "書生":82} # 辞書オブジェクトの作成
print(dict['我輩']) # キー'我輩'の値の出力
dict['我輩'] += 1 # キー'我輩'の値に1を加える
dict['柴犬'] = 1 # 新しいレコードの追加

# 全体の出力（ソート）
for key in sorted(dict):
    print(key + '\t' + str(dict[key]))
````

---

以下のように出力されたら成功です。最初の1行の数字は、キー'我輩'の値の出力です。

> `64`  
> `我輩	65`  
> `書生	82`  
> `柴犬	1`  
> `猫	75`

`sorted`メソッドを使わなかったプログラムの出力と比較してみましょう。結果が異なることが見てとれると思います。とはいうものの、実は漢字のソートは期待通りに動作しないことが多いです。熟語によって読みが異なるのが理由の一つです。しかし少なくとも出力順序は一意に定まりますので、本演習ではこれで良しとします。

これは以下のように`if-else文`を用いて処理します。

````python
if key in dict:
    # keyの値を1増やす
else:
    # 新しいレコードを追加して、値を1にする
````

---
これは、

> 1: `dict`オブジェクトに変数`token`をキーとするレコードがあり、その結果、`if`文が真を返した場合は  
> 2: この行の処理をする  
> 3: `if`文が偽を返す（キーがない）場合は  
> 4: この行の処理をする

というプログラムです。

それでは、ここまで学んだことを使って、分かち書きの結果を辞書オブジェクトに追加し、出現回数を数えながら、最後に辞書オブジェクト全体を出力するプログラムを書いてみましょう。

**ヒント1：** 辞書オブジェクト変数`dict`にはこれまでの例でデータが入っているので、`dict = {}`という処理を入れて、辞書オブジェクトを初期化するのを忘れないようにしましょう。

**ヒント2：** 辞書オブジェクトに使うキーは、`token.surface`メソッドが返す単語本体の文字列です。

````python
# janome.tokenizerから、分かち書きクラスTokenizerを読み込む
from janome.tokenizer import Tokenizer
t = Tokenizer()


# sample_doc1.txtを、openメソッドで開き、ショートカットを変数fに格納
DATA = "../data"
f = open(DATA + '/sample_doc1.txt', 'r')


# 辞書オブジェクトの初期化
dict = {}

# ファイルをforループで1行ずつ処理
for line in f:
    tokens = t.tokenize(line)
    
    for token in tokens:
        if token.surface in dict:
            dict[token.surface] += 1
        else:
            dict[token.surface] = 1

# 全体の出力（ソート）
for key in sorted(dict):
    print(key + '\t' + str(dict[key]))
````

### 複数ファイルの分かち書き
無事に、フォルダ内のファイル一覧が取得できたので、ファイル毎に分かち書きをするプログラムを作成します。

プログラムの内容が少し複雑になってきたので、処理の順番を整理しましょう。

1. フォルダに含まれるファイルの一覧を取得
1. 一覧のファイルを一つずつ開く
1. 開いたファイルを1行ずつ分かち書き
1. 分かち書きの結果を出力

という流れでいきたいと思います。

> このように、書きたいプログラムの流れを分割して整理することは、大きなプログラムを書く際にとても重要なことですので、実践していきましょう。

上記の4つの流れを実現すると以下のようになります。

````python
import os
from janome.tokenizer import Tokenizer
t = Tokenizer()

DATA = "../data"
for filename in os.listdir(DATA):
    f = open(DATA + '/' + filename, 'r')

    for line in f:
        tokens = t.tokenize(line)
    
        for token in tokens:
            print(token.surface)
````

---
上のプログラムは、

> 1-3: ライブラリの読み込み、クラスの初期化  
> 5: DATAフォルダの指定  
> 6: フォルダの一覧取得（処理1）  
> 7: ファイルを開く（処理2）  
> 9-13: ファイルを1行ずつ分かち書き処理（処理3）  
> 12-13: 分かち書きの結果を出力（処理4）

をしています。インデントも処理の入れ子を綺麗に反映してます。

これまでのプログラムと異なるのは、`open`メソッドで開くファイル名が変数`filename`に置き換わっていることのみで、それ以外は既存のプログラムを組み合わせたものです。

> とはいうものの、`for`ループが3層の入れ子になっているので、4つの処理がどのようにプログラム内で表現されているかを、注意深く理解しましょう。

それでは、上のコードセルを実行してみましょう。

すべてのファイルが分かち書きされて表示されたら成功です。

### 複数ファイルから単語の出現回数を数える

それでは、いよいよ複数ファイルから1行ずつ分かち書きをして、出現回数を数え、最終結果を表示するプログラムを作成しましょう。

````python
# 必要なライブラリを読み込む
import os
from janome.tokenizer import Tokenizer
t = Tokenizer()

# 辞書オブジェクトの初期化
dict = {}

# DATAフォルダに含まれるファイルを一つずつ処理
DATA = "../data"
for filename in os.listdir(DATA):
    f = open(DATA + '/' + filename, 'r')

    for line in f:
        tokens = t.tokenize(line)
    
        for token in tokens:
            if token.surface in dict:
                dict[token.surface] += 1
            else:
                dict[token.surface] = 1
                    
# 全体の出力（ソート）
for key in sorted(dict):
    print(key + '\t' + str(dict[key]))
    
````
